#Business Questions and Visualizations
**1.What are the top-performing product categories in terms of revenue and profit?**


```python
category_sales = df.groupby('Category').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()

sizes = category_sales['Sales']
colors = category_sales['Profit']

# Plot the treemap
mpt.figure(figsize=(10, 6))
squarify.plot(sizes=sizes, label=category_sales['Category'], color=mpt.cm.GnBu(colors/max(colors)), alpha=0.8)

mpt.title("Treemap of Product Categories by Sales and Profit")
mpt.axis('off')
mpt.show()
```

The treemap visualization will help answer one of the major business questions, which is about identifying the top categories based on revenue and profit.

Identification of high-revenue categories will help the us to understand which categories falls in high cashflow, hence prioritizing those categories for future investment or marketing campaigns.

Sales alone don't tell everything. Color-coding profit in the treemap helps the business visualize which product categories generate revenue but may not be as profitable. It helps in devising strategies for increasing either the profit margins of poor-performing categories or promoting more profitable ones.

Optimize Inventories and Marketing in cases where there is low profitability with high sales, cost management strategies or pricing need to be adopted. Products falling within high profitability can be aggressively promoted. It helps in inventory optimization, price strategy, and promotion on categories.

**2.Are there any sales trend over time, and if yes, then is there any seasonality?**


```python
# Creating new columns for year and month
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month

# Group by year and month for sales trend
sales_trend = df.groupby(['Year', 'Month']).agg({'Sales': 'sum'}).reset_index()

# Finding the month with the highest sales
top_sales_month = sales_trend[sales_trend['Sales'] == sales_trend['Sales'].max()]

# Plotting the line chart for sales trend over time
mpt.figure(figsize=(12, 6))

sb.lineplot(x='Month', y='Sales', hue='Year', data=sales_trend, palette="coolwarm", legend=False)

mpt.axvspan(top_sales_month['Month'].values[0]-0.5, top_sales_month['Month'].values[0]+0.5,
            color='yellow', alpha=0.3, label='Top Sales Month')

mpt.text(top_sales_month['Month'].values[0], top_sales_month['Sales'].values[0],
         f"Top Sales: {top_sales_month['Sales'].values[0]:,.0f}",
         horizontalalignment='center', color='red', weight='bold')

mpt.xticks(ticks=np.arange(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
mpt.grid(False)

mpt.show()
```

If we group the sales data by year and month, it can show you which way the overall trend is heading over multiple years. It highlights how sales are increasing and decreasing over the period, so that a we can mobilize its resources, predict revenue, and strategize.

The line plot shows the seasonal pattern of sales like identifying the top sales month, helping business owners to know those months with the highest number of sales in case they want to mobilize their resources during these peak months.  
* For example, every December you can see a sales boost and should staff accordingly, or conversely reduce January inventories as needed by other parts of your business.

This data allows us to drawn up the inventory management and marketing plans to be taken place in advance.

**3.Which geographical regions are the most profitable, and how do sales differ by region?**


```python
region_sales = df.groupby(['Region', 'Category']).agg({'Sales': 'sum'}).reset_index()

fig = px.sunburst(region_sales, path=['Region', 'Category'], values='Sales',
                  color='Sales',
                  color_continuous_scale='Greens',
                  title='Sales Distribution by Region and Category',
                  maxdepth=2)

fig.update_layout(margin=dict(t=40, l=0, r=0, b=0), title_x=0.5)
fig.update_traces(textinfo="label+percent entry", insidetextorientation='radial')

fig.show()
```

image.png

Here in the above code:
* By limiting the depth (maxdepth=2), we focus only on Region and Category, avoiding too many nested layers
* We use textinfo="label+percent entry", which only shows the category name and the percentage of total sales it represents. This makes the text short and easier to read.
* A monochromatic color scale (Green) is applied to reduce the complexity of the color palette ensures the chart is easier to understand. Simply put more the intensity of the colour, more the sales percentage
* The title is centered and the margins are reduced for a clean, minimalist layout that keeps the focus on the sunburst

In business context the above graph is for analyzing regional sales performance.
* The sunburst chart plots the sales information by region and category to show at once which areas are the most profitable; these could be areas a company might want to direct resources to, or attempt to optimize the worst performing.
* It also allows us to understand which products are driving the growth in a particular area. This would enable focused marketing and the creation of strategies that would apply specifically to the regional preferences aligning well with the regional market dynamics.

**4.What is the effect of discounts on profitability and sales volume?**


```python
low_profit_threshold = -1000

low_profit_df = df[df['Profit'] < low_profit_threshold]

mpt.figure(figsize=(10, 6))

# Main scatter plot: Plot all data points
sb.scatterplot(x='Discount', y='Profit', data=df, color='gray', alpha=0.5)

# Highlight low-profit points with a different color
sb.scatterplot(x='Discount', y='Profit', data=low_profit_df, color='red', s=100, label='Low Profit')

# Add regression line (trend line) and label it
sb.regplot(x='Discount', y='Profit', data=df, scatter=False, color='blue', line_kws={"linewidth": 2, "alpha": 0.8})

mpt.axhline(0, color='black', linestyle='--', alpha=0.7)

mpt.text(0.6, -3000, "Trend Line: Decreasing profit with higher discounts", color='blue', fontsize=12, weight='bold')

# Customize legend
mpt.legend(loc='upper right')

mpt.show()

# Display the list of low-profit products
print("List of Low-Profit Products with Discounts Applied:")
low_profit_list = low_profit_df[['Product Name', 'Discount', 'Profit']].sort_values(by='Profit')
display(low_profit_list)
```

Knowing the effects of discounts is critical in evaluating whether discounts lead to enough sales to compensate for the reduced profit margins or not.

Evaluating Discount Impact on Profitability by plotting the relationship between discounts and profitability, we understand how discounting affects profit margins.

The trend line (Blue line) demonstrates that higher discounts are associated with lower profits and specifically highlight products where profitability significantly impacted is (in red dots). It also guide us on the ideal discount level that maximizes overall revenue without severely impacting profitability.

**5.Which customer segment generates the most profit for the company?**


```python
# Group by customer segment and aggregate profit
segment_performance = df.groupby('Segment').agg({'Profit': 'sum'}).reset_index()

# Sorting values by profit for better visualization
segment_performance = segment_performance.sort_values(by='Profit', ascending=False)

# Create the Lollipop Chart
mpt.figure(figsize=(10, 6))
mpt.stem(segment_performance['Segment'], segment_performance['Profit'], linefmt='--', markerfmt='o', basefmt=" ", use_line_collection=True)

mpt.title("Profit by Customer Segment", fontsize=14)

mpt.grid(False)
mpt.xticks(fontsize=10)
mpt.yticks(fontsize=10)

mpt.show()
```

Customer segments profitability analysis helps us to figure out the customer segment creating the most profit for the company. This insight is key for us to maximize profitable customer engagement by targeting High-Value Customer segments that are most profitable for us through guided marketing and retention initiatives towards these groups.

This can be done by creating the Perfect Customer Retention Strategies like Optimize Marketing and Sales Efforts: Automating some or all of these calculations can free up sales and product marketing staff to focus on their strengths.  
For Example : High-profit segments can receive exclusive offers, loyalty incentives, and personalized communications to maintain customer loyalty and lifetime value and the lower-profit segments could be targeted with more cost-effective marketing campaigns.
